## Five Numer Summary

There are various descriptive statistics available, including measures such as mean, median, mode, skewness, kurtosis, standard deviation, and quartiles, each of which provides us with insights into different aspects of our data. Rather than examining these descriptive statistics in isolation, sometimes combining them can provide us with a more complete picture. In this vein, the five-number summary is a suitable method of consolidating five key descriptive statistics.

* <b>Minimum </b> it is the smallest value in our dataset.


* <b>First quartile</b> this number, Q1, corresponds to the value below which 25% of your data falls.


* <b>Median </b>  this isthe central value of your data, such that 50% of your data falls below the median.


* <b>Third quartile</b> this number, Q3, corresponds to the value below which 75% of your data falls.


* <b>Maximum</b> it is the largest value in our dataset.


The mean and standard deviation can be used together to describe both the central tendency and the spread of a dataset, but both statistics are sensitive to outliers. In contrast, the median, first quartile, and third quartile are less affected by outliers.


```python
#Example

import numpy as np
import matplotlib.pyplot as plt

data = [3,5,7,9,10,12,13,14,16,18,20,23,25,27,30]

# five number summary
minimum = min(data)
q1 = np.percentile(data,25)
median = np.median(data)
q3 = np.percentile(data,75)
maximum = max(data)

#box plot
plt.boxplot(data)
plt.title("Sales Ratio Chart")
plt.show()

print("Minimum: ",minimum)
print("Q1 : ",q1)
print("Median: ", median)
print("Q3: ",q3)
print("Maximum: ", maximum)
```

## Graphs


In statistical studies, information is often presented in the form of pictures, shapes, or lines to make it easier to understand. These visual representations are called graphs.

A graph is a method of expressing numerical data through lines. By visualizing numerical data, graphs allow for comparisons between different sets of data, making it possible to interpret and understand the information. Converting tables and charts into graphs enables the identification and evaluation of increases and decreases in data. It is important to include a unit of measurement on the vertical axis when creating a graph.

### Column Charts

Data that is arranged only in columns or rows on a worksheet can be plotted in a column chart. A column chart typically displays categories along the horizontal axis and values along the vertical axis, as seen in this type of chart.

#### Clustered Column Chart:

A clustered column chart displays values in 2D columns. Use this chart when you have categories representing:

* Value ranges (e.g., item counts)


* Specific scale arrangements (e.g., Likert scale entries containing input such as strongly agree, agree, neutral, disagree, strongly disagree)


* Names that are not in any specific order (e.g., item names, geographic names, or person names).


#### Stacked Column Chart:
A stacked column chart displays values in 2D stacked columns when you have multiple data series and want to emphasize the total.

#### 100% Stacked Column Chart:

A 100% stacked column chart displays values in 2D stacked columns that add up to 100%. Use this chart when you have two or more data series and want to emphasize their contributions to the whole, especially if the total is consistent for each category.



```python
import matplotlib.pyplot as plt
products = ["Books","Electronics","Clothes","Food","Domestic Products"]
sales = [800,700,400,600,900]
plt.bar(products, sales)
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("E-commerce Product Sales")
plt.show()
```

```python
#Example

import random as r
import numpy as np
import matplotlib.pyplot as plt

n=50

width = [r.uniform(0.1,1.5) for i in range(n)]
length = [r.uniform(0.5,3.0) for i in range(n)]
height = [r.uniform(0.2,1.0) for i in range(n)]
weight = [r.uniform(1.0,10.0) for i in range(n)]


print("Part widths: ",width)
print("")
print("Part lengths: ",length)
print("")
print("Part heights: ",height)
print("")
print("Part Weights ",weight)

gwi=np.mean(width)
gle=np.mean(length)
ghe=np.mean(height)
gwe=np.mean(weight)

swi = np.std(width)
sle = np.std(length)
she = np.std(height)
swe = np.std(weight)

#Graph

labels=["Width", "Length", "Height", "Weight"]
means = [gwi, gle, ghe, gwe]
stds = [swi, sle, she, swe]

x=[i for i, _ in enumerate(labels)]
plt.bar(x, means, stds,
         yerr = stds, align='center', alpha=0.5)
plt.xticks(x, labels)
plt.ylabel("Means")
plt.title("Part Features")
plt.show()
```

```python
#Example


import matplotlib.pyplot as plt
import numpy as np

performance = {
     'Department A': [45,25,30,50],
     'Department B': [30,40,45,35],
     'Department C': [35,35,35,40],
     'Department D': [20,30,20,25]
 }


Departments = list(performance.keys())
DepartmentScores = list(performance.values())


fig = plt.subplots(figsize=(10,5))
quartiles = len(Departments)
barwidth = 0.2
x1 = np.arange(quartiles)
x2 = [ x + barwidth for x in x1]
x3 = [ x + barwidth for x in x2]
x4 = [ x + barwidth for x in x3]


plt.bar(x1, DepartmentScores[0], color = 'r', width = barwidth, label = Departments[0])
plt.bar(x2, DepartmentScores[1], color = 'g', width = barwidth, label = Departments[1])
plt.bar(x3, DepartmentScores[2], color = 'b', width = barwidth, label = Departments[2])
plt.bar(x4, DepartmentScores[3], color = 'y', width = barwidth, label = Departments[3])


plt.xlabel("Quartiles")
plt.ylabel("Performance Scores")
plt.title("Performance of Departments in Quarters ")
plt.xticks( [ i+ barwidth for i in range(quartiles)],
            ['1st Q','2nd Q','3rd Q','4th Q'])
plt.legend()
plt.show()
```

### Line Graphs

Line graphs are suitable for displaying data arranged in columns or rows on a worksheet. In a line graph, category data is evenly distributed along the horizontal axis, while all value data is evenly distributed along the vertical axis. Line charts can display continuous data over time on a scale axis, making them ideal for showing trends in data at equal intervals, such as monthly, quarterly, or fiscal year intervals.

#### Line and Dashed Line Charts

Line and dashed line charts can be used to represent separate data values with or without markers, particularly when there are numerous data points to display and their sequential order is significant. They can depict trends over time or equally spaced categories. When there are multiple categories or the values are approximate, an unmarked line chart can be utilized.

#### Stacked Line and Dashed Stacked Line Charts

Stacked line and dashed stacked line charts, marked or unmarked, can represent individual data values by showing the contribution trend of each value over time or equally spaced categories.

#### %100 Stacked Line and Dashed %100 Stacked Line Charts

%100 stacked line and dashed %100 stacked line charts, marked or unmarked, can display the percentage contribution trend of each value over time or equally spaced categories. When there are multiple categories or the values are approximate, an unmarked %100 stacked line chart should be employed.

```python
#Example

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10,100)
y = np.sin(x)

plt.plot(x,y,color='magenta',
          linewidth=1, linestyle='-', marker='o',
          markersize=3)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Sinus Graph")
plt.grid(True)
plt.show()
```

### Pie Charts

Data arranged in a column or row within a spreadsheet can be represented in a pie chart. Pie charts illustrate the proportionate size of each data point in relation to the total sum of all points within a given data series. The data points in a pie chart are displayed as a percentage of the entire pie.

Pie charts may be preferred in the following conditions:

* You have only one data series.

* None of the values in your data are negative.

* Very few values in your data are zero.

* You have a maximum of seven categories, each representing a part of the pie.


```python
#Example
import matplotlib.pyplot as plt

portfolio = ["Stocks","Gold",'Bonds',"Real Estate",
          "Deposit"]
weight = [0.35,0.3,0.2,0.1,0.5]


plt.pie(weight, labels = portfolio, autopct='%1.1f%%')
plt.title("Portfolio Distribution")
plt.show()
```

### Histgoram

A histogram is the representation of a grouped data distribution using a column chart. Histogram allows you to visualize the frequency distribution of a dataset. Histograms are one of the most commonly used methods to plot the graph of past data. Typically, the data is displayed in a bar chart that allows viewers to quickly digest the information. The key is to present the information in a logical order. It is a simple graph that uses a horizontal and vertical axis.

```python
#Exaample

import random as r

values = [r.randint(100,1000) for i in range(0,30)]

import matplotlib.pyplot as plt

plt.hist(values, bins=10, edgecolor='black')
plt.title("Distribution of Values")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
