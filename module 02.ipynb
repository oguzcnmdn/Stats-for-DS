## Descriptive Statistics

Descriptive statistics are statistical measures that make it easy to obtain basic information about a variable, summarize data, and indicate certain characteristics of the variable, such as central tendency, dispersion, and distribution, with respect to a sample or population. These are known as summary statistics that reflect specific properties of the variable.

Descriptive statistics are typically divided into two main groups of analysis.

* Central Tendency Measures or Measures of Location


* Measures of Distribution (Measures of Dispersion/Variation)

### Central Tendency Measures

Central tendency measures are statistical measures that indicate the tendency of units wihtin a sample or population to cluster around certain values within the observation range ($X_{min} - X_{max}$) of the variable being stuied.

Central tendency measures include:

* Mean (also known as Average)


* Median


* Mode


* Quartiles


* Percentiles

### <span style='color:red'> Mean (Average) </span>

The central tendency measure known as the mean indicates the tendency of units being studied to cluster arounda certain value within the observation range   ($X_{min} - X_{max}$) of the variable being analyzed, and the direction in which the observed values of the variable tend to lean.

* <span style='color:blue'>Aritmetic mean</span>

    The aritmetic mean is often referred to simply as the 'mean'. The mean is calculated in two different ways depending on the type of data being analyzed, using different formulas:

###### <center><span style='color:green'>Mean Formula for Norganized Data (raw, ungrouped data)</span></center> 

$$ \bar{x} = \frac{\sum_{i=1}^{n}x_i}{n}{}$$

$$i = 1,2,...,n $$




```python
#Example

import statistics as stats

data = [75,83,90,68,72,78,92,86,95]

mean = stats.mean(data)

round(mean, ndigits = 2)
```

###### <center><span style='color:green'>Mean Formula for Organized Data (tabulated or grouped raw data)</span></center> 

$$\bar{x} = \frac{\sum_{i=1}^n f_i x_i}{\sum_{i=1}^n f_i}$$ 

$$i = 1,2,...,n $$

```python
#Example

import numpy as np
import matplotlib.pyplot as plt

classes = [ (-0.5,0.5),(0.5,1.5),(1.5,2.5),(2.5,3.5),(3.5,4.5) ]
frequency = [12,20,38,55,25]

midpoints = [ (class_[0]+class_[1])/2 for class_ in classes ]
weighted_sum = sum([midpoints[i]*frequency[i] for i in range(len(classes))])

total_frequency = sum(frequency)

mean = round(weighted_sum / total_frequency, ndigits = 2)

fig, ax = plt.subplots( figsize = (8,6) )

ax.hist(np.repeat(midpoints,frequency), bins = [-0.5,0.5,1.5,2.5,3.5,4.5], edgecolor = 'black')
ax.axvline(mean, color = 'red', linestyle = '--', label = 'Mean: {x}'.format(x = mean) )
ax.set_xticks([-0.5,0.5,1.5,2.5,3.5,4.5])
ax.set_xlabel('Values')
ax.set_ylabel('Frequency')
ax.legend()
plt.show()
```

### <span style='color:red'> Geometric Mean </span>

Geometric mean is a typical value calculated in cases where there values in the series follow an increasing or decreasing arithmetic or geometric sequence.

$$\bar x_G = \sqrt[n]{x_1x_2\cdots x_n} = \left(\prod_{i=1}^nx_i\right)^{1/n}$$




```python
#Example

number_of_visitors = [120,140,180,100,200,150,130,170,190,110]

import matplotlib.pyplot as plt
from scipy.stats.mstats import gmean #gmean --> Geometric Mean

gmean = gmean(number_of_visitors)

print("Geometric mean of visitors is", gmean)

plt.hist(number_of_visitors, bins = 5)
plt.axvline(x = gmean, color = 'r', label = 'Geometric Mean')
plt.legend()
plt.show()
```

### <span style='color:red'> Harmonic Mean </span>

If the values in the dataset represent a time series, and the dataset is obtained by combining the results of k experiments that were not performed at equal intervals, and consecutive numbers in the dataset exhibit fluctuations (such as monthly, seasonal, or annual fluctuations), the central tendency measure of the data is calculated using the harmonic mean.

$$\bar x_H = \frac{n}{\frac{1}{x_1}+\frac{1}{x_2}+...+\frac{1}{x_n}}=\frac{n}{\sum_{i=1}^n\frac{1}{x_i}}$$

```python
#Example

import numpy as np
import matplotlib.pyplot as plt

data = [14, 27, 41, 121, 36, 47, 105, 18, 18, 76]

hmean = len(data) / np.sum(1 / x for x in data)

print("Harmonic Mean: ", hmean)

plt.plot(data, 'bo')
plt.axhline(hmean, color = 'g', linestyle = '--')
plt.title('Patients with diagnosis')
plt.xlabel('Time')
plt.ylabel('Number of Patients')
plt.legend(['Time','Harmonic Mean'])
plt.show()
```

### <span style='color:red'> Weigthed Mean </span>

It is a type of averaging that is applied when a central tendency measure is desired to be calculated for a larger sample, using data from research conducted at different locations and times.

$$ \bar{x}_w = \frac{sum_{i=1}^{n}w_i x_i}{\sum_{i=1}^{n}w_i} $$

```python
#Example

import matplotlib.pyplot as plt

clinic = ['A', 'B', 'C', 'D']
patients = [88, 112, 56, 12]
drugs = [4.250, 3.500, 4.0, 6.25]

wsum = sum([patients[i]*drugs[i] for i in range(len(clinic))])

total_patients = sum(patients)

wmean = wsum / total_patients

print('Weighted Mean: ', wmean)

plt.bar(clinic, patients)
plt.title('Number of Patients')
plt.xlabel('Clinics')
plt.ylabel('Patients')
plt.show()

plt.bar(clinic, drugs)
plt.title('Dose of the Drugs')
plt.xlabel('Clinics')
plt.ylabel('Doses')
plt.show()
```

### <span style='color:red'> Median </span>

The value that divides a series arranged from smallest to largest into two equal parts is called the median.

If a series has extremely small or large values from the main trend, the averages may not accurately reflect the true trend due to the influence of these values.

If there are too many extreme outliers at both ends of a series, the median value becomes the statistic that represents the series.

In frequency distribution, the value that divides the series into two equal parts in terms of one unit of measurement is the median.

It is calculated differently in data sets and frequency tables. If ***'n'*** is even, the median is the average of the two middle observations.

```python
#Example

import numpy as np
import matplotlib.pyplot as plt

data = [165, 178, 170, 166, 171, 172, 177, 180, 173, 175, 175, 169]

sorted_data = sorted(data)

print('Sorted Data: ', sorted_data)

n = len(sorted_data)

print('Len of the series', n)

if n % 2 == 0:
    median = (sorted_data[n // 2 - 1] + sorted_data[n // 2]) / 2
else:
    median = sorted_data[n // 2]

##print('Median', median)

#or (simpe way)

print('Median:', np.median(data))

plt.hist(sorted_data, bins = range(164,181,2), edgecolor = 'b')
plt.axvline( x = median, color = 'r', linestyle = 'dashed', linewidth = 2)
plt.title('Distribution of Height')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
#Median values in the frequency tables

data = [55, 60, 65, 70, 75, 80, 85, 90, 95, 100]

table = {
    'Lower Limit': [50, 60, 70, 80, 90],
    'Upper Limit': [59, 69, 79, 89, 99],
    'Frequency': [10, 15, 20, 18, 7]
    }

sorted_data = sorted(data)

n = sum(table['Frequency'])

median_index = n // 2
cumulative_value = 0
for i in range(len(table['Frequency'])):
    cumulative_value += table['Frequency'][i]
    if cumulative_value >= median_index:
        lower_limit = table['Lower Limit'][i]
        value = table['Frequency'][i]
        width = table['Upper Limit'][i] - table['Lower Limit'][i]
        median = lower_limit + ((median_index - (cumulative_value - value)) / value) * width
        break

print('Median: ', median)
```

### <span style='color:red'> Mode </span>

The most frequently recurring value in a series is known as the mode. If a series has multiple values repeated a similar number of times, it is called a bimodal or multimodal series. In tabulated data, the mode is located in the class with the highest frequency. This distribution is referred to as a peaked or modal distribution.

```python
#Example

data = [165, 170, 166, 171, 172, 177, 180, 173, 175, 175, 169]

table = {}

data.sort()

for v in data:
    if v in table:
        table[v] += 1
    else:
        table[v] = 1

print('table:', table)        
        
maxF = max(table.values())

mode = [k for k, v in table.items() if v == maxF]

print('Mode', mode)

#or (simple way)

import numpy as np
from scipy import stats

arr = np.array(data)

print('Mode', stats.mode(arr))
```

```python
# Mode in the frequency tables

import matplotlib.pyplot as plt

data = [130, 135, 140, 145, 150, 155,160, 170, 175, 180, 185, 190, 195, 200]

table_ = {
    'Lower Limit': [130, 140, 150, 160, 170, 180, 190],
    'Upper Limit': [139, 149, 159, 169, 179, 189, 199],
    'Frequency': [5, 9, 12, 8, 5, 3, 1]
    }

index = table_['Frequency'].index(max(table_['Frequency']))
mode = (table_['Lower Limit'][index] + table_['Upper Limit'][index]) / 2

print('Mode:', mode)

plt.bar(table_['Lower Limit'], table_['Frequency'], width = 10, edgecolor = 'black')
plt.xlabel('Borders')
plt.ylabel('Frequency')
plt.title('Car Data')
plt.axvline(mode, color = 'r', linestyle = 'dashed', linewidth = 3, label = 'Mode')
plt.show()
```

### <span style='color:red'> Quartiles </span>

The first quartile, denoted by Q1, represents the 25th percentile of a dataset, indicating the value below which 25% of the data falls.

The second quartile, denoted by Q2, is also known as the median and represents the 50th percentile of the dataset, indicating the value below which 50% of the data falls.

The third quartile, denoted by Q3, represents the 75th percentile of the dataset, indicating the value below which 75% of the data falls.

```python
#Example

import random
import matplotlib.pyplot as plt

data = [random.randint(0, 100) for i in range (20)]

fig, ax = plt.subplots()
ax.boxplot(data, vert = False)
plt.show()
```

### <span style='color:red'> Percentiles </span>

Percentiles are statistics that divide a sorted dataset into equal hundredths. They are denoted as P(%), where P(5) represents the 5th percentile of the dataset, indicating the value below which 5% of the data falls.

### Measures of Distribution

Distribution measures, also known as descriptive statistics, are used to determine the distribution, dispersion of values within a range, and concentration tendencies at certain values of a variable.

Distribution measures are measures that support central tendency measures and show the distribution, spread, and dispersion around the given central tendency measures.

### <span style='color:red'> Varience</span>

A measure of dispersion that describes how the data in an array are distributed around the mean and the state of spread and scatter as the square of the variable's measure unit.

$$\operatorname{Var}(X) = \frac{1}{n} \sum_{i=1}^n (x_i - \bar{x})^2$$



```python
#Example

import random
import numpy as np
import matplotlib.pyplot as plt

data = [random.randint(1, 10) for i in range (20)]

mean = np.mean(data)

variance = sum((x - mean) ** 2 for x in data) / len(data)
print('Variance', variance)

plt.hist(data, edgecolor = 'b' )
plt.xlabel('Survey Score')
plt.ylabel('Frequency')
plt.title('Educaiton Satisfaction Survey')
plt.show()
```

### <span style='color:red'> Standard Deviation </span>

A statistic used to describe how the data in a series is distributed around the mean, and the spread and dispersion.

The unit of measure is the same as the unit of measure of the variable. It is denoted by the letter $\sigma$.

$$\sigma = \sqrt{\frac{\sum_{i=1}^{n}(x_i - \mu)^2}{n}}$$


```python
#Example

import random
import statistics as stats

data = [random.randint(1, 20) for i in range (20)]

n = len(data)

std = stats.stdev(data)
print('Standard Deviation', std)
```

### <span style='color:red'> Standard Error </span>

The standard error is a measure of how consistent the sample mean is in estimating the population mean, and how much error is carried by the estimate. It is obtained by dividing the standard deviation by the square root of the sample size.

$$SE = \frac{\sigma}{\sqrt{n}}$$

```python
#Example

import random
import statistics as stats

data = [random.randint(1, 20) for i in range (20)]

n = len(data)

std = stats.stdev(data)

se = std / (n**0.5)
print('Standard Error', se)
```

The standard deviation provides information about the distribution of values in a dataset, while the standard error provides information about the distribution of sample means that can be obtained from a sample of size n.

In scientific publications, it is important to make generalizations about the population, and therefore it is necessary to provide the sample mean with its standard error.

### <span style='color:red'> Coefficient of Variation </span>

The coefficient of variation is a distribution measure used to compare the variability of variables with different unit numbers and measurement units to their mean.

It is calculated using the mean and standard deviation of the variable

$$CV = \frac{\sigma}{\mu} \times 100\%$$
