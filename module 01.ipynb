## What is Statistics?

Statistics is a discipline that involves the collection of data for a specific purpose, summarizing the data using tables and graphs, interpreting the results, describing the level of confidence in the findings, generalizing the results obtained from samples to be population, investigating relationships between variables, making predictions about various topics related to the future, and designing experiments and observations.

It is based on the gathering, categorizing, analyzing, and interpretation of data to achieve a specific objective.

Statistics can be divided into two branches:

### <span style='color:red'> 1. Descriptive Statistics </span>

Descriptive statistics can be used to smmarize a sample numerically or graphically. Examples of numerical measures unclude the mean and standart deviation. Graphical summaries may include various types of charts and tables.

Examples of numerical measures include:

* Measures of location (arithmetic mean, geometric mean, harmonic mean, quadratic mean, mode and median)
* Measures of dispersion (range, variance, standard deviation, etc.)
* Measures of skewness and kurtosis

Descriptive statistics are useful for describing the important characteristics of a dataset, such as the central tendecny and variability of the data. These measures can help in summarizing and interpreting the data, making it easier to understand and communicate the findgins to others.

### <span style='color:red'> 2. Inferential Statistics </span>

Inferential statistics is used to mode relationships in the data, taking into account probability, and to draw conclusions about a larger statistical population. These conclusions can take the form of yes/no answers (hypothesis testing), estimates of numerical characteristics (statistical estimation), predictions of future values (statistical forecasting), interpretation of linear relationships between variables (correlation), or modeling of these relationsships (regression analysis). Other commony used mathematical modeeling techniques include variance analysis, ANOVA, time series anaylsis, and data mining.

Inferential statistics build on the foundations of sampling theory and probabily theory, and is used to make inferences about a population based on a sample of data. Hypthesis testing is used to determine wheter and observed effect is statistically significant, while regression and corretlation analysis are used to study the relationships between variables.

Inferential statistics can be used to make predictions and draw conclusions about future events, as well as to test the validity of scientific hypotheses. It is a powerful tool for decision-making and problem-solving in a wide range of fields, including business, medicine, engineering, and social sciences.

### The Subject of Statistics

The subject matter of statistics can be divided into two categories based on the ability of events to represent their own type of events accurately. These categories are called typical events and collective events.

Typical events are identical events that repeat in the same way when necessary conditions are met. Physical and chemical events are examples of typical events. As these events are identical to each other, only on of them can represent the groupy they form. The information of water through hydrogen and oxygen under certain conditions is a typical event.

Collective events, on the other hand, are vents that have common aspects but are not identical and have differences among them. They are generally named after events related to living organisms. Population is a good example of a collective event. Althought the individuals that make up the population have comon characteristics such as being human and living in the same region or country, they differ in various characteristics such as gender, age, and occupation.

In summary, statistics deal with both typical events, where the events are identical and can represent the whole, and collvetive events where the are common aspects but not necessarily identical and cannot be represented by a single event.

### <span style='color:red'>Variable: </span>

A variable is any characteristics or attribute that can take on different values or categories in a population or sample. These values can be either quantitative or qualitative, and they may be discrete or continous.

**Quantitative** variables are numeric and can be measured on a scale, such as heigh, weight, or age. They can be further categorized as either discrete or continous. **__Discrete variables__** are those that can only take on specific, separate values, such as the number of children in a family. **__Continous variables__**, on the other hand can take on any value within a certain range, such as time, temperature, or distance.

**Qualitative variables**, also knows as categorical variables, do not have a numerical value and are classifed into different categories. For example, gender, marital status, or educational level.

A variable is a characteristics that can vary across different units in a population or sample, and understanding the nature of the variable is crucial in statistical anallysis as it determines the type of analysis that can be performed on the data.

### <span style='color:red'>Measurement Level: </span>

In statistics, the measurement level is a concept that determines the type of a variable's measured quality. The measurement level of a variable level of a variable determines which statistical operations can be applied to it.

There are four basic categories of measurement levels in statistics:

* <span style='color:blue'>Nominal Scale:</span> In this scale, the variable consist of categorical or qualitative categories. Examples of variables include gender, blood type, or eye color.


* <span style='color:blue'>Ordinal Scale:</span> In this scale, the variable contains orderable and categorical data. However, differences or ratios between the data cannot be calculated. Examples of variables include edication level, satisfaction level, or ranking.


* <span style='color:blue'>Interval scale:</span> In this scale, differences between the data can be calculated, but ratios cannot be calculated. Examples of variables include temperature, year, or hour.


* <span style='color:blue'>Ratio scale:</span> In this scale, differences and ratios between variables can be calculated. Examples of variables include income, population, or height.


The measurement level determines which statistical operations can be applied to variables. For example, for a nominal scale variable, only mode, percentage, and frequency calculations can be performed. However, for ratio scale variables, arithmetic mean, standard deviation and percantage change calculations can be performed.

### <span style='color:red'>Series: </span>

A series is used to group and classify a dataset that requires the establisment of boundaries. A series can comprise a set of classes used to group data within a particular range or class intervals. The series may also include the upper and lower limits of the classes, as well as their widths. Additionally, the series may indicate the frequency or percentage distribution of the classes.

For instance, a statistical dataset may include the grades of students. In this case, a series can be created to classify and group the grades of the students. The series can be developed by grouping the student grades into specific ranges or classifications. This process makes the data more meaningful and easily understandable.

Overall, the creation of a series is a fundamental technique used in statistical analysis. By organizing data in a systematic and logical manner, it becomes possible to gain a deeper understanding of the underlying patters and trends present in the data. Thus, the use of a series is a crucial tool in making informed decisions based on statistical data.

* <span style='color:blue'>Distribution series:</span> Distribution series, also known as frequency distribitution, is an ordered set of data that has been classified into classes or intervals and shows the frequency of the occurence of each class or intervals. Distribution series are used to analyze the distribution of data and identify patterns and trends.

    The process of creating a distribution series involves sorting data into classes or intervals based on a predetermined range or grouping. The classes or intervals are typically defined by their lower and upper limits and their width. The frequency of occurence of each class or interval is then determined, and this information is used to construct a frequency distribution table or chart.

    Distributon series provide a visual representation of the distribution of data and allow for easy identification of patterns and trends. They re an essential tool for statistical analysis and can help researchers and professionals make informed decisions based on the distribution of data.
    

* <span style='color:blue'>Simple series:</span> A simple series is a type of statistical series where data is grouped into specific intervals or classes, and the frequency or count of data points within each interval is calculated. The intervals are typically defined by the range of the data and the desired width or size of each interval.

    Simple series be useful for simmarizing a large data set in a compact and meaningful way. It can help to identify patterns or trends within the data and make it easier to interpret and analyze. Additionally, a simple series can be used to create graphical representations of the data, such as histograms or frequency polygons, which can provide a visual understanding of the data distribution.
    
    
* <span style='color:blue'>Cumulative series:</span> A cumulative series is a type of statistical series where the frequency of each class is calculated by adding up the frequencies of all classes below it. This means that each class frequency represents the total number of observations that fall within that class and all previous classes.

    To create a cumulative series, we start witha simple series, which is a series of data that has been grouped into speciic intervals or classes, and the frequency of data points within each interval has been calculated. We then add up the frequencies of each class, starting from the first class and continuing to each subsequent class, to get the cumlative frequency.
    
    
* <span style='color:blue'>Compound series:</span> A compound series, also known as a composite series, is a type of statistical series that is created by combining two or more individual series. This is done by grouping the data from each series into common intervals or classes, and then combining the frequencies of each interval or class across al the individual series.

    Compound series are useful when we want to analyze data from multiple sources or for different subgroups within a larger population. By combining the data into a single series, we can identify patterns or trends that might not be apparent when looking at each individual series seperately.
    
    Compoun series can also be used to analyze data from multiple time periods or regions. For example, we might create individual frequency series of each year in a time series data set, and then combine them into a compound series to identify long-term trends or seasonal patterns.


* <span style='color:blue'>Density series:</span> A density series, also known as a probability density function, is a type of statistical series that represents the probability distribution of a continous random variable. Unlike a frequency series, which is used to represent the distribution of a discrete random variable, a density series is used for variables that can take on any value within a certaing range.

    The are under the curve in a density series represents the probability that the random variable falls within a certain range. This can be calculated by intergrating the probability density function over the desired range.
    
    Density series are commonly used in statistical analysis to model and analyze continous data, such as measurements of physical quantities, financial data, or other continous variables. ### <span style='color:red'>Frequency tables: </span>

Frequency tables are a type of data table used to classify and group statistical data. A frequency table sghows the number or percentage of different classes in a data set.

A frequency table enables data to be represented in a classified and grouped manner. For example, the grades of students in a class can be thought of as a data set. The grades can be grouped into a certain range and presented as a frequency table. This table can show the frequency or percentage of each grade cass.

The structure of a frequency table can include different characteristics such as **__class range, lower and upper limits, class width, and class midpoint__**.These features are important for the classification and grouping of data and should be claculated accurately.


$$Class\ Range = Maximum\ Value - Minimum\ Value $$



$$Class\ Width = \frac{Range}{Number\ of\ Classes}$$



$$ Lower\ Limit = (Class\ Boundaries)\ - \frac{1}{2}\ (Class\ Width) $$



$$ Class\ Midpoint = \frac{Lower\ Limit + Upper\ Limit}{2} $$


Frequency tables provide a meaningful way to present statistical data and are a fundemental tool for data analysis. They allow for a more understandable visualization of classified and grouped data and enable easier comparison of data.

There are different types of frequency measures that can be used in statistical analysis. Here are some of the most common ones:

* <span style='color:blue'>Absolut frequency:</span> This refers ti the actual number of times a particular value occurs in a data set.


* <span style='color:blue'>Relative frequency:</span> This is the proportion of the total number of observations that belong to a particular value. It is calculated by dividing the absolute frequency by the total number of observations.


* <span style='color:blue'>Cumulative frequency:</span> This is the sum of the frequencies of all values up to a certain point in the data set. For example, the cumulative frequency of a ceratin value is the sum of the absoulte frequencies of that value and all values before it.


* <span style='color:blue'>Cumulative relative frequency:</span> This is the proportion of the total number of observations that fall below a certain value. It is calculated by dividing the cumulative frequency by the total number of observations.


* <span style='color:blue'>Percentage frequency:</span> This is the frequency expressed as a percentage of the total number of observations. It is calculated by multiplying the relative frequency by 100.


* <span style='color:blue'>Percentage cumulative frequency:</span> This is the cumulative frequency expressed as a percentage of the total number of observations. It is calculated by multiplying the cumulative relative frequency by 100.


### <span style='color:red'>Central Tendency Measures: </span>

Central tendency measures are statistics that describe the central or typical value of a set of observations. They are used to provide a summary of the data and to make comparisons between different data sets.

* <span style='color:blue'>Aritmetic mean:</span> It is also known as the average, is a measure of central tendency that is calculated by summing up all the values in a data set and dividing the result by the total number of obervations. It is the most commonly used measure of central tendency and is sensitive to extreme values, which can significantly affect the result.

$$\text{Aritmetic Mean} = \bar{x} = \frac{\sum_{i=1}^{n}x_i}{n}$$

* <span style='color:blue'>Geometric mean:</span> It is a measure of central tendency that is used to find the average value of a set of numbers that are multiplied together. It is typically used to calculate growth rates and investment returs.

    The geometric mean is calculated by taking the nth rooth of the product of n numbers, where n is the number of values in the data set. The formula for culculating the geometric mean is:
    
$$\text{Geometric Mean} = \sqrt[n]{x_1x_2\cdots x_n} = \left(\prod_{i=1}^nx_i\right)^{1/n}$$

The geometric mean is useful when dealing with data sets that exhibit exponential growth, such as population growth or compund interst rates. It is also used in finance to calculate the average rate of return on an investment over multiple periods of time.

* <span style='color:blue'>Harmonic mean:</span> Harmonic mean is a measure of central tendency that is used to find the average of a set of numbers that are rates or ratios. It is calculated by dividing the total number of obervations by the sum of the reciprocals of each observation.

$$\text{Harmonic Mean} = H = \frac{n}{\frac{1}{x_1}+\frac{1}{x_2}+...+\frac{1}{x_n}}=\frac{n}{\sum_{i=1}^n\frac{1}{x_i}}$$

* <span style='color:blue'>Square mean:</span> The square mean, also known as the root mean square, is a statistical measure that represents the square root of the arithmetic mean of the squares of a set of values. It is commonly used to calculate the average of a set of numbers that may include both positive and negative values. The square mean is a way to measure the magnitude of a set of numbers.

To calculate the square mean, you first square each number in the set of values, then take the arithmetic mean of the resulting set of squares, and finally take the square root of this mean. The formula for the square mean is:

$$\text{Square Mean} = \sqrt{\frac{x_1^2 + x_2^2 + \cdots + x_n^2}{n}}$$

* <span style='color:blue'>Weighted mean:</span> It is a statistical measure that takes into account the relative importance or wight of each value in a data set. The weighted mean is calculated by multiplying each value by its corresponding weight, adding up the values, and the dividing the total by the sum of the weights. This type of mean is commonly used when different values in a data set have different levels of significance or importance. For example, in calculating the overall grade of a student, the grades on different assignments may be weighted differently based on their relative importance.

$$\text{Weighted Mean} = \bar{x}_w = \frac{sum_{i=1}^{n}w_i x_i}{\sum_{i=1}^{n}w_i} $$

* <span style='color:blue'>Median:</span> The median is the value that seperates a dataset into two equal halves when all values are arrenged in order. It is the middle value in a sorted list of values. The median is not affect by outliers in the dataset and can be ysed as measure of central tendency in a skewed distribution.

    The formula for finding the median depends on the number of observations in the dataset. If the number of observations **'n'** is odd, the median is the middle observation. If **'n'** is even, the median is the average of the two middle observations.
    
    
* <span style='color:blue'>Mode:</span> It is the value that appears most frequently in a dataset. If a dataset has one mode, it is considered unimodal. If it has two modes, it is bimodal, and if it has three or more modes, it is multimodal.

    Modality is an important concept in statistical analysis because it provides information about the shape of a distribution. For example, an unimodal distribution with a single peak is often described as "bell-shaped" and is charecteristic of a normal distribution. A bimodal distribution with two peaks might suggest that the data come from two different populations.
    
    
* <span style='color:blue'>Quartiles:</span>  Quartiles are values that divde an ordered dataset into four equal parts. The three quartiles divide the dataset into quartes, with the first quartile (Q1) representing the 25th percentile, the second quartile (Q2) representing the 50th percentile (which is also the median), and the third quartile (Q3) representing the 75th percentile.

    The quartiles are used in many statistical calculartions, such as the interquartile range (IQR) and box plots. The IQR is the distance betweem the first and third quartiles and provides a measure of the spread of the middle 50% of the data. Box plots use the quartiles to display the distribution of data graphically.
